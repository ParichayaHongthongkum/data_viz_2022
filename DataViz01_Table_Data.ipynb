load data

```python
import pandas as pd  #ทำงานกับข้อมูลลักษณะตาราง
import os  #จัดการไฟล์และ path
```

```python
from google.colab import drive  #เชื่อม google drive
drive.mount('/content/drive')
path = '/content/drive/My Drive/midterm data 2022'   #pathจะชี้ไปยังโฟล์เดอร์
```

```python
data_file_path = os.path.join(path,'confirmed-cases.xlsx') #แล้วก็ชี้ไปที่ไฟล์
print(data_file_path)
```

```python
data_covid = pd.read_excel(data_file_path)  #กำหนดตัวแปร data_covid เป็นข้อมูลของผู้ป่วยโควิด
data_covid  #เรียกดูตารางของผู้ป่วยโควิด
```

เลือกมาเฉพาะคอลัมน์ที่ต้องการจะวาดกราฟ

```python
data_covid_select = data_covid[['sex','age','risk']]
data_covid_select
```

ตรวจสอบค่า Missing

```python
data_covid_select.isnull().any()  # .isnull() ดูค่าว่าง ถ้ามีจะแสดงเป็น  true และจากการแสดงผลพบว่าข้อมูลมีค่า Missing อยู่
```

แก้ค่า missing

```python
#แทนค่า Missing ของคอลัมน์ sex ด้วยค่า Mode (ฐานนิยม)
data_covid_select['sex'].mode()  ## หาค่าที่ซ้ำกันมากที่สุดโดยใช้ .mode
```

```python
data_covid_select=data_covid_select.fillna(value ={ 'sex' : 'หญิง'})  ## แก้คอลลัมน์ sex เติมค่าว่าง โดยใช้คำสั่ง .fillna() ซึ่งค่าที่นำมาใส่คือ หญิง
data_covid_select
```

```python
set(data_covid_select['sex']) # ใช้ set ในการดูว่าในคอลัมน์ sex มีการจัดกลุ่มอย่างไร พบว่า มีการจัดกลุ่ม ที่ถูกแบ่งออกเป็น ชาย , นาย , หญิง
```

```python
data_covid_select[data_covid_select['sex']=='นาย'] #เมื่อทำการแสดงข้อมูล พบว่ามีเพียงข้อมูลเดียว จึงเดาว่าน่าจะมีการผิดพลาดของข้อมูล
```

```python
data_covid_select= data_covid_select.replace('นาย','ชาย') # ใช้ .replace ในการเปลี่ยนค่าข้อมูลในตาราง จาก ชาย ให้เป็น นาย
```

```python
set(data_covid_select['sex']) # เมื่อที่การเซตเพื่อดูการจัดกลุ่มพบว่า เหลือแค่ค่าที่เป็นชายและหญิงเท่านั้น
```

```python
#แทนค่า Missing ของคอลัมน์ age ด้วยค่า Mean (มัธยฐาน)
data_covid_select['age'].mean()  ## หาค่ากลางของอายุโดยใช้ .mean()
```

```python
data_covid_select=data_covid_select.fillna(value ={ 'age' : data_covid_select['age'].mean()})  ## แก้คอลลัมน์ age เติมค่าว่าง โดยใช้คำสั่ง .fillna() ซึ่งค่าที่นำมาใส่คือ ค่าจากค่ากลางของอายุ
data_covid_select
```

```python
#แทนค่า Missing ของคอลัมน์ risk ด้วยค่า Mode (ฐานนิยม)
data_covid_select['risk'].mode()  ## หาค่าที่ซ้ำกันมากที่สุดโดยใช้ .mode
```

```python
data_covid_select=data_covid_select.fillna(value ={ 'risk' : 'สัมผัสใกล้ชิดกับผู้ป่วยยืนยันรายก่อนหน้านี้'})   ## แก้คอลลัมน์ risk เติมค่าว่าง โดยใช้คำสั่ง .fillna() ซึ่งค่าที่นำมาใส่คือ สัมผัสใกล้ชิดกับผู้ป่วยยืนยันรายก่อนหน้านี้
data_covid_select
```

```python
data_covid_select.isnull().any()  #ตรวจสอบค่า Missing
```

### เลือก class ที่จะนำมาใช้ในการวาดกราฟ

```python
data_covid_select['risk'].unique() # ดูการจัดกลุ่มของข้อมูล
```

**technique :** https://blog.datath.com/cheatsheet-pandas/#withi_leuxk_hlay_khxlamn_cak_DataFrame

```python
# เลือก class ที่ต้องการ
data_class = data_covid_select[(data_covid_select['risk']=='บุคลากรด้านการแพทย์และสาธารณสุข')| (data_covid_select['risk']=='พิธีกรรมทางศาสนา')|(data_covid_select['risk']=='Cluster บ่อนไก่อ่างทอง')] 
data_class
```

```python
data_class['risk'].unique() # ดูการจัดกลุ่มของข้อมูล
```

เนื่องจากจะวาดกราฟ จึงต้องแปลงค่า sex ให้เป็นตัวเลข

```python
data_class= data_class.replace('ชาย',0)  #แทนค่าของผู้ชาย เป็นเลข 0
data_class= data_class.replace('หญิง',100) # แทนค่าของผู้หญิง เป็นเลข 70
data_class
```

```python
pd.plotting.parallel_coordinates(data_class,'risk',color = ['#F5DEB3','#E50000','#008000']);
```

เนื่องจาก class ไม่สามารถแสดงเป็นภาษาไทยได้ จึงเปลี่ยนเป็นภาษาอังกฤษ

```python
data_class= data_class.replace('บุคลากรด้านการแพทย์และสาธารณสุข','medical personnel')
data_class= data_class.replace('พิธีกรรมทางศาสนา','religious ritual')
data_class= data_class.replace('Cluster บ่อนไก่อ่างทอง','Cluster Angthong Casino')
data_class
```

```python
pd.plotting.parallel_coordinates(data_class,'risk',color = ['#F5DEB3','#E50000','#008000']);
```

จากกราฟ  
* ทั้งชายและหญิง ที่มีความเสี่ยงจากการเป็นบุคคลากรทางการแพทย์ พบว่าอยู่ในช่วงอายุใกล้เคียงกัน
* กลุ่มที่มีความเสี่ยงจากพิธีกรรมทางศาสนา ส่วนใหญ่จะเป็นผู้ชาย ซึ่งโดยอายุส่วนใหญ่มากกว่า 60 ปี ขึ้นไป
* กลุ่มที่มีความเสี่ยงมาจากบ่อน มีทั้ง ชาย และหญิง โดย หญิงส่วนใหญ่จะอายุต่ำกว่า 20 ปี และชายส่วนใหญ่จะอายุ 40 ปีขึ้นไป
