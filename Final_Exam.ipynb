##เชื่อมตาราง

```python
alldb_drop # ตารางข้อมูล alldb
```

```python
hospital_choose # ข้อมูล hospital
```

```python
data_all01= alldb_drop.merge(hospital_choose,how='left',left_on='site',right_on='hcode') # เชื่อมตาราง ทั้งสองตารางเข้าด้วยกัน โดยใช้ ตัวเชื่อมคือ site และ hcode 
data_all01
```

```python
data_all01.isnull().sum() #ตรวจสอบค่าว่างของข้อมูล
```

```python
data_all02 = data_all01.dropna() # ลบค่า Missing ของข้อมูลโดยการใช้ dropna()
data_all02
```

```python
removes = data_all01.shape[0] - data_all02.shape[0]
print(f'size before drop = {data_all01.shape[0]}')
print(f'size after drop = {data_all02.shape[0]}')
print(f'we lost {100*(removes/data_all02.shape[0])}% of data')
```
