````python
this_data = data_covid[['sex','age','province_of_onset']]
````

````python
male = this_data[this_data['sex']=='ชาย']
male['age']
````

````python
sum = 0
N = 0
for a in male['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ชาย {sum/N}')
````

````python
male_KK = male[male['province_of_onset']=='ขอนแก่น']

sum = 0
N = 0
for a in male_KK['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1
print(f'อายุเฉลี่ย ของ ผู้ชาย ในจังหวัดขอนแก่น {sum/N}')        
````

3.หาจำนวนผู้ป่วยที่เป็นคน ที่ไม่ใช่คนไทย

````python
data_covid  ###แสดงตาราง data_covid เพื่อดูชื่อ column ของ ตาราง data_covid
````

````python
Ex3 = data_covid[data_covid['nationality']!='Thailand'] ##กำหนด Ex3 ให้เลือกข้อมูลในตาราง data_covid ที่ column nationality ไม่เป็น Thailand

N = 0 ##กำหนด N เป็น 0 เพื่อใช้นับจำนวนผลลัพธ์ใน loop
for a in Ex3['No.']: ## loop ในตาราง Ex3 ที่ column 'No.'
    
    if a > 0: #ถ้าตาราง Ex3 ที่ column 'No' มากกว่า 0 ให้ :
        N += 1 #ให้ N บวกเพิ่มขึ้น 1
print(f'จำนวนผู้ป่วยที่เป็นคน ที่ไม่ใช่คนไทย {N} คน') ### แสดงข้อมูล
````

## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021 

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021



````python
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````

### HW กลุ่ม 2 || เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20  

````python
this_data = data_covid[['announce_date']] #กำหนดชื่อData แล้วเลือกชื่อcolumnที่ต้องการ
````

````python
this_data #แสดงData
````

````python
TF=list() #กำหนดTF=listว่าง
for each_row in this_data.iterrows(): #สร้าง list เพื่อวนลูปของแต่ละแถว
    if each_row[1]['announce_date'].split('/')[1] <= '20':  #ในแต่ละ1แถวใช้คำสั่งsplit เพื่อแบ่งstringข้อความออกเป็นส่วนๆ โดยใช้ตัวคั่นคือ /
        TF.append(True)
    else:
        TF.append(False)

this_data[TF]
````
